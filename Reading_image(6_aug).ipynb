```python
#the image has been correctly loaded by opencv as a umpy array but the color of each pixel has been sorted as BGR . matplotlib plot expects an rgb image so for a correct display of image  it is necessary to swap those channels
this operation can be done either by using opencv conversion functions cv2.cvtcolor() or by working directly with the numpy array
```
